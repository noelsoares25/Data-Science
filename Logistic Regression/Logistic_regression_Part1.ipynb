# Data wrangling/Cleaning

```python
titanic_dataset.isna().any()
```

```python
titanic_dataset.isna().sum()
```

```python
titanic_dataset.drop('Cabin',axis=1,inplace=True)
```

```python
titanic_dataset.isna().sum()
```

```python
titanic_dataset.dropna(inplace=True)
```

```python
titanic_dataset.isna().sum()
```

```python
titanic_dataset.Sex
```

```python
sex=pd.get_dummies(titanic_dataset['Sex'],drop_first=True)
sex.head(10)
```

```python
embarked=pd.get_dummies(titanic_dataset['Embarked'],drop_first=True)
embarked.head(10)
```

```python
pcl=pd.get_dummies(titanic_dataset['Pclass'],drop_first=True)
pcl.head(10)
```

```python
titanic_dataset=pd.concat([titanic_dataset,sex,embarked,pcl],axis=1)
```

```python
titanic_dataset.head()
```

```python
titanic_dataset.drop(['PassengerId','Name','Sex','Pclass','Ticket','Embarked'],axis=1,inplace=True)
```

```python
titanic_dataset.head(10)
```

## Test, Train split

```python
X=titanic_dataset.drop(['Survived'],axis=1)
y=titanic_dataset['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=4)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
X_train
```

```python
model.fit(X_train,y_train)
```

```python
predictions=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,predictions)
```

```python
y_test.value_counts()
```

```python
confusion_matrix(y_test,predictions)
```
